## 2. Data Cleaning and Preprocessing

In this stage, we will prepare our data for analysis and modeling. This includes:
1. Converting categorical column ('Yes'/'No') to numerical format (1/0).
2. Dropping unnecessary columns that do not contribute to the prediction.

```python
# Convert "Internship_Experience" and "Placement" to numerical format
df['Internship_Experience'] = df['Internship_Experience'].map({'Yes': 1, 'No': 0})
df['Placement'] = df['Placement'].map({'Yes': 1, 'No': 0})

print("Categorical columns converted successfully!")
```
